# chapter 1. 헬로 파이썬
## 1.1 파이썬이란?
파이썬은 간단하고 배우기 쉬운 프로그래밍 언어\
파이썬은 과학 분야, 특히 인공지능과 데이터 과학 분야에서 널리 쓰임\
파이썬 자체의 뛰어난 성능에 넘파이와 사이파이 같은 수치 계산과 통계 처리를 다루는 탁월한 라이브러리 존재\
텐서플로, 파이토치, 케라스 같은 유명 딥러닝 프레임워크들이 파이썬용 API 제공

## 1.2 파이썬 설치하기
### 1.2.1 파이썬 버전
파이썬은 2.x와 3.x라는 두 가지 버전이 공존\
버전의 하위 호환성이 없으므로 어느 버전을 설치할지 선택\
여기에서는 파이썬 3를 사용함

### 1.2.2 사용하는 외부 라이브러리
* 넘파이 : 수치 계산용 라이브러리 <br> 넘파이에는 고도의 수학 알고리즘과 배열(행렬)을 조작하기 위한 편리한 메서드가 준비되어 있음
* 맷플롯립 : 그래프를 그려주는 라이브러리 <br> 맷플롯립을 이용하면 실험 결과를 시각화하거나 딥러닝 실행 과정의 중간 데이터를 시각적으로 확인할 수 있음

### 1.2.3 아나콘다 배포판
파이썬을 설치하는 여러 방법 중 아나콘다라는 배포판을 이용하기를 권함\
배포판이란 사용자가 설치를 한 번에 수행할 수 있도록 필요한 라이브러리 등을 한 묶음으로 정리해둔 꾸러미\
그중 아나콘다는 데이터 분석에 중점을 둔 배포판

## 1.3 파이썬 인터프리터
파이썬을 설치했다면 파이썬 버전을 먼저 확인\
터미널에서 python --version 명령어를 실행


```python
# ipynb에서 터미널 명령어 실행 
!python --version
```

위처럼 'Python 3.x.x'가 표시된다면 파이썬 3가 제대로 설치된 것\
이어서 python이라고 입력하여 파이썬 인터프리터를 시작

```python
!python
```

파이썬 인터프리터는 '대화 모드'라 하여, 개발자와 파이썬이 대화하듯 프로그래밍 할 수 있음\
'대화하듯'이라는 말은, 예를 들어 개발자가 '1+2는?'이라고 물으면 파이썬 인터프리터가 '3'이라고 대답함

### 1.3.1 산술 연산
\*은 곱셈, \/는 나눗셈, \**는 거듭제곱을 의미하며, 파이썬 3에서는 정수를 나눈 결과는 실수(부동소수점)

```python
1-2
```

```python
4*5
```

```python
7/5
```

```python
3**2
```

### 1.3.2 자료형
프로그래밍 언어에는 자료형(data type)이 존재함\
자료형이란 데이터의 성질을 나타내는 개념\
예를 들어 정수, 실수, 문자열과 같은 형태가 존재함\
파이썬에는 type()함수로 특정 데이터의 자료형을 알아볼 수 있음

```python
type(10)
```

```python
type(2.718)
```

```python
type('hello')
```

### 1.3.3 변수
x와 y 등의 알파벳을 사용하여 변수(variable)를 정의할 수 있음\
변수를 사용하여 계산하거나 변수에 다른 값을 대입할 수 있음

```python
x = 10
print(x)
```

```python
y = 3.14
type(x*y)
```

파이썬은 동적 언어로 분류되는 프로그래밍 언어\
동적이라 함은 변수의 자료형을 상황에 맞게 자동으로 결정한다는 뜻\
위의 예제에서 x의 자료형이 정수임을 사용자가 명시한 적이 없지만 10이라는 정수로 초기화 할 때, x의 형태가 정수임을 파이썬이 스스로 판단함\
또한, 정수와 실수를 곱한 결과는 실수가 되었음(자동 형변환)

### 1.3.4 리스트
여러 데이터를 리스트(list)로 정리할 수 있음

```python
a = [1,2,3,4,5]
print(a)
```

```python
len(a)
```

```python
a[0]
```

```python
a[0]=99
print(a)
```

원소에 접근할 때는 a[0]처럼 작성\
`[ ]`안의 수를 인덱스(색인)라 하며 인덱스는 0부터 시작함\
파이썬 리스트에는 슬라이싱(slicing)이라는 편리한 기법이 존재하며, 슬라이싱을 이용하면 범위를 지정해 원하는 부분 리스트를 얻을 수 있음

```python
print(a)
```

```python
a[0:2]
```

```python
a[1:]
```

```python
a[:3]
```

```python
a[:-1]
```
